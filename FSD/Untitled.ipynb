```python
import pandas as pd
import numpy as np
```

```python
fil_infl=pd.read_csv('filtered_influencer_profiles.csv')
sent_fil=pd.read_csv('sentiment_final_filtered.csv')
```

```python
print(fil_infl.columns.tolist())
```

```python
fil_infl.drop(columns=['url','tf','profile_pic','profile_pic_missing','combined_text','follow_ratio','posts_per_follower','follow_ratio_scaled','posts_scaled','engagement_scaled'],inplace=True)
```

```python
print(fil_infl.columns.tolist())
print(sent_fil.columns.tolist())
```

```python
# Merge based on username
final_filtered_influencer = pd.merge(
    fil_infl,
    sent_fil,
    on="username",
    how="left"     # keeps all influencers even if some have no sentiment data
)

print("âœ… Influencer + Sentiment data merged successfully!")

# Save to CSV
final_filtered_influencer.to_csv("final_filtered_influencer.csv", index=False)

print("ðŸ’¾ Saved as final_filtered_influencer.csv")
```
