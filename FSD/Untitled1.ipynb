```python
# Make a copy so we don't mess original
sent_summary = sent_df.copy()

# Total comments
sent_summary['total_comments'] = sent_summary['positive_comments'] + sent_summary['neutral_comments'] + sent_summary['negative_comments']

# Percentages
sent_summary['positive_pct'] = (sent_summary['positive_comments'] / sent_summary['total_comments']) * 100
sent_summary['neutral_pct']  = (sent_summary['neutral_comments']  / sent_summary['total_comments']) * 100
sent_summary['negative_pct'] = (sent_summary['negative_comments'] / sent_summary['total_comments']) * 100
```

```python
sent_summary['sentiment_score'] = (sent_summary['positive_pct'] * 1 + 
                                   sent_summary['neutral_pct']  * 0.5 + 
                                   sent_summary['negative_pct'] * 0) / 100 * 100
```

```python
sent_summary = sent_summary.groupby('username', as_index=False)[
    ['positive_pct', 'neutral_pct', 'negative_pct', 'sentiment_score']
].mean()
```
