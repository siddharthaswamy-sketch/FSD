```python
import pandas as pd

# --- Step 1: Aggregate sentiment counts per influencer ---
agg_sentiment = sentiment_df2.groupby("username")[["positive_comments", "neutral_comments", "negative_comments"]].sum().reset_index()

# --- Step 2: Calculate total comments ---
agg_sentiment["total_comments"] = (
    agg_sentiment["positive_comments"] +
    agg_sentiment["neutral_comments"] +
    agg_sentiment["negative_comments"]
)

# --- Step 3: Calculate sentiment percentages ---
agg_sentiment["positive_percentage"] = round((agg_sentiment["positive_comments"] / agg_sentiment["total_comments"]) * 100, 2)
agg_sentiment["neutral_percentage"]  = round((agg_sentiment["neutral_comments"] / agg_sentiment["total_comments"]) * 100, 2)
agg_sentiment["negative_percentage"] = round((agg_sentiment["negative_comments"] / agg_sentiment["total_comments"]) * 100, 2)

# --- Step 4: Merge with sentiment_score_df ---
sentiment_final_df = pd.merge(
    agg_sentiment,
    sentiment_score_df[["username", "sentiment_score", "dominant_sentiment"]],
    on="username",
    how="left"
)

# --- Step 5: Optional: keep only relevant columns ---
sentiment_final_df = sentiment_final_df[
    ["username", "positive_percentage", "neutral_percentage", "negative_percentage", "sentiment_score", "dominant_sentiment"]
]

print("‚úÖ sentiment_final_df created successfully!")
sentiment_final_df.head()
```
